```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
df = pd.merge(movies, ratings)

movie_ratings = pd.pivot_table(df, index='userId', columns='title', values='rating')
movie_ratings.head(2)
```

```python
reetings = movie_ratings.reindex(movie_ratings.count().sort_values(ascending=False).index, axis=1)
reetings.head(3)
```

```python
reetings_replaced = reetings.fillna(0)
reetings_replaced.head(3)
```

```python
# K-Means clustering
km = KMeans(n_clusters=5)
X = reetings_replaced 
km = km.fit(X)
```

```python
reetings['Cluster'] = km.predict(X)
reetings.head(20)
```

```python
reetings["Cluster"].value_counts()
```

there seem to be 3 important main clusters

```python
r_pivot = reetings.reindex(reetings.count().sort_values(ascending=False).index, axis=1)
r_pivot.tail(5)
```

```python
mean_per_cluster = r_pivot.groupby(r_pivot["Cluster"]).mean()
mean_per_cluster
```
