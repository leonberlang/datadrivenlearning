```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv("iris.csv")
iris.tail(10)
```

```python
sns.scatterplot(x="PetalLengthCm", y="SepalLengthCm", hue="Species", data=iris)
```

```python
sns.scatterplot(x="PetalLengthCm", y="SepalLengthCm", style="Species", data=iris)
```

```python
# K-Means clustering
km = KMeans(n_clusters=3)
X = iris[['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] 
km = km.fit(X)
```

```python
# adding new predicted cluster to dataframe
iris['Cluster'] = km.predict(X)
iris.tail(20)
```

```python
sns.scatterplot(x="PetalLengthCm", y="SepalLengthCm", style="Species", hue="Cluster", data=iris)
```

Let's try 2 clusters.

```python
# K-Means clustering
km = KMeans(n_clusters=2)
X = iris[['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] 
km = km.fit(X)
```
