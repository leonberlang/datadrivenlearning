```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("studentdata.csv")
df.tail(10)
```

```python
def remove_trash(x):
    if(x == "IOS"): 
        return ('iOS')
    else: 
        return x

df['os_phone'] = df['os_phone'].apply(remove_trash)
```

```python
df.tail(10)
```

```python
os_totals = df["os_phone"].value_counts()
os_totals
```

```python
print("2 versies, 1: pandas plot versie en 2: seaborn plot versie")
os_totals.plot(kind="bar")
sns.catplot(x="os_phone", kind="count", palette="magma_r", data=df);
```

```python
percentages = pd.crosstab(df.gender, df.os_phone, normalize=True)
percentages
```

```python
print("BAR GRAPH - NOT STACKED - count can only have one axis, seperated by hue...")
sns.catplot(x="gender", hue="os_phone", kind="count",
            palette="winter", edgecolor=".6",
            data=df);
```
