```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("class_data.csv")
```

```python
# calculating mean and median 
mean_size = df['instagram_followers'].mean()
median_size = df['instagram_followers'].median()
print(f'mean: {mean_size}')
print(f'median: {median_size}')

# commentary
print("ok.. we have to kill some outliers")
```

```python
# stripping everyone above 500 followers
df_subset = df[df['instagram_followers']<500]

# calculating mean and median again
mean_size = df_subset['instagram_followers'].mean()
median_size = df_subset['instagram_followers'].median()
print(f'mean: {mean_size}')
print(f'median: {median_size}')
```

```python
# plotting an histogram
df_subset['instagram_followers'].name = 'Instagram followers (amount)' #Labeling the variable for plotting
amount_followers = df_subset['instagram_followers'].dropna() #Selecting the size column. Dropping NA's, otherwise SNS won't plot
sns.distplot(amount_followers, kde=False) # kde = density plot eventueel erdoorheen 
plt.title('Distribution Instagram followers')
plt.show()
```

```python
# Plotting app distribution from classmates in a violin plot
df_apps = df[df['apps_phone'] < 1000] #Subsetting the dataframe to include only more than 5 apps
sns.violinplot(df_apps['apps_phone'])
sns.swarmplot(df_apps['apps_phone'], color='white') #default color doesn't stand out
plt.title('Total apps distribution')
plt.ylabel('Students')
plt.xlabel('Amount of apps')
plt.show()

# Calculating mean and median (also to visualize in next distribution plot)
mean_size = df_apps['apps_phone'].mean()
median_size = df_apps['apps_phone'].median()
print(f'mean: {mean_size}')
print(f'median: {median_size}')

# Plotting distribution plot, with density line (no kde=False included)
sns.distplot(df_apps['apps_phone'])
plt.title('Total apps distribution')
plt.ylabel('Amount of people')
plt.xlabel('Total apps on phone')
plt.axvline(median_size, 0, 100, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_size, 0, 100, color='blue', label='mean')
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
# Showing total texts distribution in a boxplot, with some cleaning
df_texts = df[df['texts_yesterday'] < 500]
sns.boxplot(df_texts['texts_yesterday'])
plt.title('Distribution total texts yesterday under classmates')
plt.xlabel('Amount of texts sent')
plt.show()
```

```python
# Plotting linkedin connections in density plot (with less than 1000 connections, to kill outliers)
df_con = df[df['linkedin_con'] < 1000]
sns.distplot(df_con['linkedin_con'])
plt.title('Distribution of LinkedIn connections')
plt.ylabel('amount of people?')
plt.show()
```
