```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr
```

```python
df = pd.read_csv("funda-amsterdam.csv")
```

```python
df.tail(3)
```

```python
sns.pairplot(df)
plt.show
```

```python
df_new = df[["area","rooms","bedrooms","price"]]
sns.pairplot(df_new)
plt.show
```

```python
df_areaprice = df[["area","price"]]
```

```python
sns.pairplot(df_areaprice)
```

```python
df_areaprice_clean = df_areaprice[df_areaprice["price"] > 10000]
```

```python
sns.pairplot(df_areaprice_clean)
```

```python
sns.scatterplot(data=df_areaprice_clean, x="price", y="area")


print("pretty linear, a few outliers!")
```

```python
pearsons_r = pearsonr(df_areaprice_clean['price'],df_areaprice_clean['area'])
r_value = list(pearsons_r)[0]
p_value = list(pearsons_r)[1]
```

There is a significant linear correlation between housing area and price! 

```python
print(f"Pearsons r: {r_value} with a p-value of {p_value}")
```
