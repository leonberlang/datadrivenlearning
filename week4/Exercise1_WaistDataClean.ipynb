```python
import pandas as pd
```

```python
steps = pd.read_csv("steps.csv", encoding="utf-8", delimiter=";")
survey = pd.read_csv("survey.csv", encoding="utf-8")
```

```python
steps_survey = pd.merge(survey,steps, on="id")
```

```python
steps_survey
```

```python
visual = steps_survey.plot(y="waist")
```

```python
highest_values = steps_survey.sort_values(by="waist",ascending=False)
highest_values.head(10)
```

```python
lowest_values = steps_survey.sort_values(by="waist", ascending=True)
lowest_values.head(10)
```

```python
# removing extreme values
def remove_extremes(x):
    if(x > 135.0): 
        return float('NaN')
    if(x < 40.0): 
        return float('NaN')
    else: 
        return x

steps_survey['waist_corrected'] = steps_survey['waist'].apply(remove_extremes)
```

```python
waist_corrected = steps_survey.plot(y="waist_corrected")
```
